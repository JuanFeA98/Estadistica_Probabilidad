## **Análisis gráfico**

### **Correlaciones**

```python
sns.heatmap(iris.corr(), cmap='PuBu', annot=True)
plt.show()
```

```python
sns.pairplot(data=iris)
plt.show()
```

```python
plt.figure(figsize=(8,6))

plt.title('Relación entre el largo del petalo y el largo del sépalo', fontweight=1000)
sns.scatterplot(data=iris, x='sepal_length' , y='petal_length', hue='species')

plt.grid()

plt.show()
```

```python
p = sns.jointplot(data=iris, x='sepal_length' , y='petal_length', hue='species')
p.fig.suptitle('Relación entre el largo del petalo y el largo del sépalo', fontweight=1000)
p.fig.tight_layout()
plt.show()
```

```python
p = sns.lmplot(data=iris, x='sepal_length' , y='petal_length', hue='species')
p.fig.suptitle('Relación entre el largo del petalo y el largo del sépalo', fontweight=1000)
p.fig.tight_layout()
plt.show()
```

### **Distribuciones**

```python
sns.boxplot(data=iris, x='species' , y='sepal_length')
plt.title('Distribución del largo del sépalo', fontweight=1000)
plt.show()
```

```python
sns.boxplot(data=iris, x='species' , y='petal_length')
plt.title('Distribución del largo del petalo', fontweight=1000)
plt.show()
```
