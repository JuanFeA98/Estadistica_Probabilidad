## **Datos**

```python
df = sns.load_dataset('iris')
```

```python
df.sample()
```

```python
p = sns.pairplot(data=df, hue='species')

plt.suptitle('Relacin entre las variables', fontweight=1000, fontsize=13)

p.fig.tight_layout()

plt.show()
```

### **Matriz de covarianza**

```python
scaler = StandardScaler()
scaled = scaler.fit_transform(df.drop('species', axis=1))
```

```python
covariance_matrix= np.cov(scaled.T)
```

```python
sns.heatmap(covariance_matrix, annot=True, fmt='.2f', annot_kws={'size': 10})

plt.title('Matriz de covarianza')

plt.show()
```

### **Matriz de correlacin**

```python
sns.heatmap(df.corr(), annot=True)

plt.show()
```
