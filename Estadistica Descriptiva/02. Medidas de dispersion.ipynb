## Librerias

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Datos

```python
iris = sns.load_dataset('iris')
```

```python
iris.sample()
```

```python
iris.describe()
```

```python
sns.heatmap(iris.corr(), cmap='PuBu')
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(data=iris, x='sepal_length' , y='petal_length', hue='species')
plt.show()
```

```python
sns.jointplot(data=iris, x='sepal_length' , y='petal_length', hue='species')
plt.show()
```

```python
sns.lmplot(data=iris, x='sepal_length' , y='petal_length', hue='species')
plt.show()
```

```python
sns.boxplot(data=iris, x='species' , y='sepal_length')
plt.show()
```

```python
sns.barplot(data=iris, x='species', y='sepal_length')
```
